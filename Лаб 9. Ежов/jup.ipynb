```python
import pandas as pd
import random
import matplotlib.pyplot as plt
import math as math
```

```python
passengers = pd.read_csv('./titanic.csv')
passengers
```

```python
a_s = passengers[(passengers["Age"] < 40) & (passengers["Age"] > 25) & (passengers["Sex"] == 'male')]
a_s
```

```python
a_s = passengers[(passengers["Survived"] == 1) & (passengers["Pclass"] == 1) | (passengers["Sex"] == 'female') & (passengers["Age"] < 25)]
a_s
```

```python
a_s = passengers[["Sex","Pclass"]]
a_s = a_s.groupby(["Sex", "Pclass"]).size()
a_s.plot.bar()
```

```python
a_s = passengers[(passengers["Age"] < 25)]
a_s = a_s[["Sex", "Pclass", "Survived"]]
a_s = a_s.groupby(["Sex", "Pclass", "Survived"]).size()
a_s.plot.bar()
```
